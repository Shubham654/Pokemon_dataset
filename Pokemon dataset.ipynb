```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pokemondb.net/pokedex/all' 
data = requests.get(url)
soup = BeautifulSoup(data.text,'lxml')
print(soup.prettify)
```

```python
name = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('a', class_ = 'ent-name')

poke_name = []
for i in range(len(name)):
    poke_name.append(name[i].text)
poke_name  
```

```python
total = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-total')

poke_total = []
for i in range(len(total)):
    poke_total.append(total[i].text)
poke_total
```

```python
hp = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_hp = []
for i in range(1 , len(hp),7):      #jump after 6                                 
    poke_hp.append(hp[i].text)
poke_hp  
```

```python
attack = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_attack = []
for i in range(2 , len(attack),7):      #jump after 6                                
    poke_attack.append(hp[i].text)
poke_attack 
```

```python
defence = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_defence = []
for i in range(3 , len(defence),7):      #jump after 6                                
    poke_defence.append(hp[i].text)
poke_defence 
```

```python
sp_atck = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_sp_atck = []
for i in range(4 , len(sp_atck),7):      #jump after 6                                
    poke_sp_atck.append(sp_atck[i].text)
poke_sp_atck
```

```python
sp_deff = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_sp_deff = []
for i in range(5 , len(sp_deff),7):      #jump after 6                                
    poke_sp_deff.append(sp_deff[i].text)
poke_sp_deff
```

```python
speed = soup.find('div',class_ = 'grid-col span-md-12 span-lg-10').find_all('td', class_ = 'cell-num')
poke_speed = []
for i in range(6 , len(speed),7):      #jump after 6                                
    poke_speed.append(speed[i].text)
poke_speed
```

```python
poke_data = {'Name':poke_name,
             'Total':poke_total,
             'HP':poke_hp,
             'Attack':poke_attack,
             'Defence':poke_defence,
             'Sp_attack':poke_sp_atck,
             'Sp_defence':poke_sp_deff,
             'Speed':poke_speed}
```

```python
pokemon_data = pd.DataFrame(poke_data)
```

```python
pokemon_data.to_csv('pokemon_df.csv',index=False)
```
